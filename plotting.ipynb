Librerías a utilizar:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Carga de los *csv*:

```python
results_1 = pd.read_csv('training_steps_1.csv')
results_1.head()

results_2 = pd.read_csv('training_steps_2.csv')
results_2.head()
```

Limpieza de columnas innecesarias:

```python
del results_1['episode']
del results_1['success']
del results_1['step']

del results_2['episode']
del results_2['success']
del results_2['step']
```

Generación de la gráfica:

```python
from matplotlib.pyplot import figure
figure(figsize=(12, 6), dpi=160)

r1 = results_1.rolling(window=10000).mean()
r2 = results_2.rolling(window=10000).mean()

plt.plot(r1, color="orange", label="XXXX", linewidth=1.0, alpha=0.4)
plt.plot(r2, color="green", label="XXXX", linewidth=1.2)

plt.legend()
plt.title('Algoritmo: XXX  | Entorno: YYY | Comparativa: ZZZZ', fontsize=10)

plt.show()
```
